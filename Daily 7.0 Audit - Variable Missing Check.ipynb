```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
import re
```

```python
b1 = pd.read_csv(r'\\neptune\RAD\1 Projects\2021\Oct\DailyAudit_YZH\Export\B1_SC_EXPORT_V7_10-22-21_13788.csv', sep='\t', encoding='cp1252')
b1.head()
```

```python
b1_col = list(b1.columns)

r = re.compile("[A-Z]{3}[0-9]{4}")
exp_col = list(filter(r.match, b1_col)) 

exp_col.extend(('VANTAGE_V4_SCORE','FICO','OPEN DATE'))
print(exp_col)

len(exp_col)
```

```python
ln_col = ['ADDRCHANGECOUNT60MONTH',
'ADDRCURRENTLENGTHOFRES',
'ADDRCURRENTTIMEOLDEST',
'ADDRINPUTSUBJECTCOUNT',
'ADDRINPUTTIMEOLDEST',
'ADDRONFILECOUNT',
'ADDRPREVIOUSTIMENEWEST',
'ADDRPREVIOUSTIMEOLDEST',
'AUTO_SCORE',
'CONFIRMATIONINPUTDOB',
'CROSSINDUSTRYSCORE',
'CROSSINDUSTRY_SCORE',
'DEROGSEVERITYINDEX',
'INPUTPROVIDEDPHONE',
'INQUIRYTELCOM12MONTH',
'SOURCECREDHEADERTIMENEWEST',
'SOURCEVOTERREGISTRATION',
'SSNPROBLEMS',
'SUBJECTACTIVITYINDEX03MONTH',
'OPEN DATE'
]

ft_col = ['NUMBEROFAPPLICATIONSALLTIME',
'NUMOFAUTOAPPLICATIONSALLTIME',
'CC2_AD',
'FTHIT',
'FTSCORE',
'LL_NO_OF_NEW_LOANS_ORGNTD',
'LL_OL_NO_OF_APPLS_90',
'LL_SF_NO_OF_APPLS',
'LL_SF_NO_OF_MERCHANTS',
'LL_SF_TTL_OS_BALANCE',
'LL_SSN_NO_OF_ADDRS',
'SFNUMBEROFLOANSPASTDUE',
'OPEN DATE'
]

app_col = list()
for col in b1_col:
    if (col not in exp_col) & (col not in ln_col) & (col not in ft_col):
        app_col.append(col)
app_col.append('OPEN DATE')        

print(app_col)
len(app_col)
```

```python
# Function to Check Null Count and Null %

def null_check(df, col_lst, grouping_var):
    gp = df[col_lst].groupby(df[grouping_var])
    gp_key = gp.groups.keys()
    missing_cnt = []
    missing_cnt_pct = []

    for i in gp_key:
        s1 = gp.get_group(i).isnull().sum()
        s1.name = i
        missing_cnt.append(s1)    
        s2 = round(gp.get_group(i).isnull().sum()/len(gp.get_group(i))*100,2)
        s2.name = i
        missing_cnt_pct.append(s2)
    
    missing_cnt_df = pd.DataFrame(missing_cnt).transpose()
    missing_pct_df = pd.DataFrame(missing_cnt_pct).transpose()
    missing_cnt_df.drop(missing_cnt_df.tail(1).index,inplace=True)
    missing_pct_df.drop(missing_pct_df.tail(1).index,inplace=True)
    return missing_cnt_df,missing_pct_df
```

```python
# Initiate lists of dataframes to export
dfs_app = []
dfs_ln = []
dfs_exp = []
dfs_ft = []
```

```python
# Application Variables Check 
app_missing_cnt,app_missing_pct = null_check(b1, app_col, 'OPEN DATE')
app_missing_cnt.name = 'Application Variables Null Count'
app_missing_pct.name = 'Application Variables Null %'
dfs_app.extend((app_missing_cnt,app_missing_pct))
print(app_missing_cnt)
print(app_missing_pct)
```

```python
# LexisNexis Variables Check 
ln_missing_cnt,ln_missing_pct = null_check(b1, ln_col, 'OPEN DATE')
ln_missing_cnt.name = 'LexisNexis Variables Null Count'
ln_missing_pct.name = 'LexisNexis Variables Null %'
dfs_ln.extend((ln_missing_cnt,ln_missing_pct))
print(ln_missing_cnt)
print(ln_missing_pct)
```

```python
# Experian Variables Check on Non-Missing Population
NonMissingB1 = b1[b1['FICO'].notnull()]
exp_missing_cnt1,exp_missing_pct1 = null_check(NonMissingB1, exp_col, 'OPEN DATE')
exp_missing_cnt1.name = 'Experian Variables Null Count on Non-Missing Population'
exp_missing_pct1.name = 'Experian Variables Null % on Non-Missing Population'
dfs_exp.extend((exp_missing_cnt1,exp_missing_pct1))
print(exp_missing_cnt1)
print(exp_missing_pct1)
```

```python
# Experian Variables Check on All Population
exp_missing_cnt2,exp_missing_pct2 = null_check(b1, exp_col, 'OPEN DATE')
exp_missing_cnt2.name = 'Experian Variables Null Count on All Population'
exp_missing_pct2.name = 'Experian Variables Null % on All Population'
dfs_exp.extend((exp_missing_cnt2,exp_missing_pct2))
print(exp_missing_cnt2)
print(exp_missing_pct2)
```

```python
# FactorTrust Variables Check on FT Hit Population
FtHitB1 = b1[b1['FTHIT']=='Y']
ft_missing_cnt1,ft_missing_pct1 = null_check(FtHitB1, ft_col, 'OPEN DATE')
ft_missing_cnt1.name = 'FactorTrust Variables Null Count on FT Hit Population'
ft_missing_pct1.name = 'FactorTrust Variables Null % on FT Hit Population'
dfs_ft.extend((ft_missing_cnt1,ft_missing_pct1))
print(ft_missing_cnt1)
print(ft_missing_pct1)
```

```python
# FactorTrust Variables Check on All Population
ft_missing_cnt2,ft_missing_pct2 = null_check(b1, ft_col, 'OPEN DATE')
ft_missing_cnt2.name = 'FactorTrust Variables Null Count on All Population'
ft_missing_pct2.name = 'FactorTrust Variables Null % on All Population'
dfs_ft.extend((ft_missing_cnt2,ft_missing_pct2))
print(ft_missing_cnt2)
print(ft_missing_pct2)
```
